```python
# 전처리된 데이터 파일 다시 불러오기
real_movie = pd.read_csv("movie.csv", index_col=0)
real_movie.head(10)
```

```python
real_movie.dtypes
```

```python
real_movie['개봉일'] = pd.to_datetime(real_movie['개봉일'])
```

```python
real_movie['연도'] = real_movie['개봉일'].dt.year
```

```python
real_movie
```

```python
real_movie.dtypes
```

```python
real_movie['매출액'] = real_movie['매출액'].str.replace(',', '').astype(float)
#real_movie['매출액'].str.replace(",","").astype(int)
```

```python
real_movie.to_csv("movie(+매출액).csv", encoding=0)
```

```python
# 2004 ~ 2007 삭제 
drop_year = real_movie[real_movie['연도'] < 2008].index
real_movie = real_movie.drop(drop_year).reset_index(drop=True)
real_movie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_year_term = real_movie['연도'].value_counts().sort_index()
movie_year_term
```

```python
korea_movie = real_movie[real_movie['국적']=='한국']
korea_year_term = korea_movie['연도'].value_counts().sort_index()
```

```python
plt.figure(figsize=(13,10))
sns.barplot(korea_year_term.index, korea_year_term.values, alpha=0.8)
plt.title('한국 연도별 영화수',fontsize=15)
plt.xlabel('year', fontsize=12)
plt.ylabel('count',fontsize=12)
plt.show()
```

```python
# 일1)막대그래프 위에 스크린 수 선그래프로 나타내기
# 일 2) 나라별로 영화개수 확인해보기; 가장 영화산업이 발달된 나라 찾기
```

```python
real_movie['스크린수'] = pd.to_numeric(real_movie['스크린수'], errors='coerce')
```

```python
# 스크린 수 
screan = real_movie.groupby(real_movie['연도']).sum()['스크린수']
screan
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.bar(movie_year_term.index, movie_year_term.values, color='g')
ax2 = ax1.twinx()
ax2.plot(screan.index,screan.values, color='r')
ax2.grid(False)
plt.show()
```
